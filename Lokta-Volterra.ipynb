```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
alpha = 1
beta = 0.5
omega = 0.5
delta = 2
```

```python
r = np.array([2,2], float)#x,y
def lokta(r,t):
    x = r[0]
    y = r[1]
    fx = alpha*x - beta*x*y
    fy = omega*x*y - delta*y
    return np.array([fx,fy],float)
```

```python
a = 0
b = 30
N = 1000
h = (b-a)/N

tpoints = np.arange(a,b,h)#begin, end, step size
xpoints = []
ypoints = []

for time in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    k1 = h*lokta(r,time)
    k2 = h*lokta(r+0.5*k1,time+0.5*h)
    k3 = h*lokta(r+0.5*k2,time+0.5*h)
    k4 = h*lokta(r+k3,time+h)
    r = r + (k1+2*k2+2*k3+k4)/6

plt.plot(tpoints, xpoints)
plt.plot(tpoints, ypoints)
plt.xlabel('Time')
plt.ylabel('Position')
plt.show()
```
