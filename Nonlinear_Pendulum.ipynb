```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
theta0 = 0.99*np.pi
g = 9.8
l = 0.1
```

```python
r = np.array([160,-2.05], float)#theta,omega
def pend(r,t):
    theta = r[0]
    omega = r[1]
    ftheta = omega
    fomega = -(g/l)*np.sin(theta)
    return np.array([ftheta,fomega],float)
```

```python
a = 0
b = 5
N = 10000
h = (b-a)/N

tpoints = np.arange(a,b,h)#begin, end, step size
thetapoints = []
omegapoints = []

for time in tpoints:
    thetapoints.append(r[0])
    omegapoints.append(r[1])
    k1 = h*pend(r,time)
    k2 = h*pend(r+0.5*k1,time+0.5*h)
    k3 = h*pend(r+0.5*k2,time+0.5*h)
    k4 = h*pend(r+k3,time+h)
    r = r + (k1+2*k2+2*k3+k4)/6

initial_theta = 160   
linear = initial_theta*np.cos(np.sqrt(g/l)*tpoints)

plt.plot(tpoints, thetapoints, 'r')
plt.plot(tpoints, linear, 'g--')
plt.xlabel('Time')
plt.ylabel('Theta')
#plt.ylim(145,165)
plt.show()
```
