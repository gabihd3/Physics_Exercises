```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Parameter definition
w=4
#Inital conditions
pos = 1 
vel = 1
#Function Definition
def f(r,t): 
    pos = r[0]
    vel = r[1] 
    fpos = vel
    fvel = (-w**2)*(pos**3)
    return np.array([fpos,fvel],float)
#Setting up the steps to use in the 4th Runge Kutta Method
a = 0
b = 1000 
N = 1000
h = (b-a)/N
tpoints = np.arange(a,b,h)
pospoints = []
velpoints = []
```

```python
#Runge Kutta Method
r = np.array([0,1],float)

for t in tpoints:
    pospoints.append(r[0])
    velpoints.append(r[1])
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)
    r = r + (k1+2*k2+2*k3+k4)/6

plt.plot(tpoints, pospoints)
plt.show()
```
