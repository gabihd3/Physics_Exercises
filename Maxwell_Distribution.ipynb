```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

Section a)

```python
kb = 1.38*(10**-23) #Boltzmann's constant

def f(v,m,T): #Function definition passing v, m and T arguments 
    return 4*np.pi*((m/2*np.pi*kb*T)**(3/2))*(v**2)*np.exp((-m*v**2)/2*kb*T) # Eq (1)
```

Section b)

```python
#CONSTANTS
T1 = 300
T2 = 1000
mhe = 6.65*(10**-27)
mn2 = 4.65*(10**-26)
#VARIABLE
v = np.linspace(0,100) 
```

i)

```python
# Lists with 100 values for a gas of He at both at T = 300 and T = 1000 respectively
He_T1 = f(v,mhe,T1) 
He_T2 = f(v,mhe,T2)
```

```python
He_T1_plot = plt.plot(He_T1,v)
plt.title('Distribution at 300K', fontsize = '20')
plt.xlabel('v', fontsize = '20')
plt.ylabel('Speed Distribution', fontsize = '20')
plt.show
```

```python
He_T2_plot = plt.plot(He_T2,v)
plt.title('Distribution at 1000K', fontsize = '20')
plt.xlabel('v', fontsize = '20')
plt.ylabel('Speed Distribution', fontsize = '20')
plt.show
```

# Simpson's Rule

```python
v1 = 0 #a
v2 = 200 #b
N = 100
```

```python
def simpson(f,v1,v2,N):
    h = (v2-v1)/N
    s = (f(v1,mhe,T1)+f(v2,mhe,T1))
    for k in range(1,N,2):
        s = s + 4*f(v1 + k*h,mhe,T1)
    for k in range(2,N,2):
        s = s + 2*f(v1 + k*h,mhe,T1)
    simp = (h/3)* s
    return simp
```

```python
simpson(f,v1,v2,N)
```
